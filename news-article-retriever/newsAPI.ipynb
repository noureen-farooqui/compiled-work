```python
import requests
import json
import pandas as pd
```

```python
api_key = 'd5396daf19364378939e5518f2b3090a'
endpoint = 'https://newsapi.org/v2/top-headlines'
categories = ['business', 'entertainment', 'general', 'health', 'science', 'sports', 'technology']
```

```python
articles = []
for category in categories:
    parameters = {
        'country': 'in',  # Change this to your desired country
        'category': category,
        'apiKey': api_key,
        'pageSize': 100 # The number of results to return per page (request). 20 is the default, 100 is the maximum.
    }

    response = requests.get(endpoint, params=parameters)
    news_data = response.json()

    for article in news_data['articles']:
        title = article['title']
        description = article['description']
        url = article['url']
        published_at = article['publishedAt']
        image_url = article['urlToImage'] if 'urlToImage' in article else None  # Check if image URL is available

        articles.append({
            'Title': title,
            'Description': description,
            'URL': url,
            'Published At': published_at,
            'Category': category,
            'Image URL': image_url
        })
```

```python
df = pd.DataFrame(articles)
df.head()
```

```python
df.to_csv('newsAPI_articles.csv')
```
