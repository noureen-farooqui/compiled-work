```python
import requests
import pandas as pd

api_key = '411664428886e0f3a42ef38e67343f38'
categories = ['general', 'world', 'nation', 'business', 'technology', 'entertainment', 'sports', 'science', 'health']

articles = []

for category in categories:
    url = f"https://gnews.io/api/v4/top-headlines?category={category}&lang=en&country=in&token={api_key}&max=20"

    response = requests.get(url)
    news_data = response.json()

    if response.status_code == 200:
        if "articles" in news_data:
            for article in news_data['articles']:
                title = article['title']
                description = article['description']
                url = article['url']
                published_at = article['publishedAt']

                articles.append({
                    'Title': title,
                    'Description': description,
                    'URL': url,
                    'Published At': published_at,
                    'Category': category
                })
        else:
            print(f"No articles found for category '{category}'")
    else:
        print(f"Error fetching data for category '{category}': {response.status_code} - {response.reason}")

# Create a DataFrame from the article data
df = pd.DataFrame(articles)
df.head()
```

```python
import requests
import pandas as pd

# Set up your API key and search engine ID
API_KEY = "AIzaSyAfZrvmTnmsteEl7_6U2yBZjw_cg3ueq9s"
SEARCH_ENGINE_ID = "b27891aaeb2ec402f"

# Function to perform the search and return results as a DataFrame
def google_search(query, category=None):
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "key": API_KEY,
        "cx": SEARCH_ENGINE_ID,
        "q": query,
        "searchType": "image" if category == "image" else "web",
        "siteSearch": category if category != "image" else None
    }
    response = requests.get(url, params=params)
    data = response.json()

    # Extract search results
    if "items" in data:
        search_results = data["items"]
        results_list = []
        for result in search_results:
            title = result["title"]
            link = result["link"]
            snippet = result["snippet"]
            image_link = result.get("image", {}).get("thumbnailLink", None)
            results_list.append({"Title": title, "Link": link, "Snippet": snippet, "Image Link": image_link})

        # Create DataFrame from search results
        df = pd.DataFrame(results_list)
        return df

    return pd.DataFrame()  # Return an empty DataFrame if no results found

# Perform a search for articles in the data science category
query = "data science"
category = "example.com"  # Specify the desired category or website for articles
search_results_df = google_search(query, category)

# Print the DataFrame
print(search_results_df)
```
