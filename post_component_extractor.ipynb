```python
import pandas as pd
import re
import demoji
```

```python
df = pd.read_csv('Book1.csv')  ## in this dataset in places of where the image occurs in the post their path has been saved
df.head()
```

```python
def extract_post_components(post):
    text_content = ""
    links = []
    image_content = []
    hashtags = []
    emojis = []

    if isinstance(post, str):
        
        ## extracting text content
        cleaned_post = re.sub(r"(?:\@|https?\://)\S+", "", post)  ## removing links
        cleaned_post = re.sub(r"\#[a-zA-Z0-9]+\b", "", cleaned_post)  ## removing hashtags
        cleaned_post = demoji.replace(cleaned_post, "")  ## removing emojis
        text_content = cleaned_post.strip()

        ## extracting links
        link_pattern = r"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
        links = re.findall(link_pattern, post)

        ## extracting image path
        image_pattern = r'(?:[a-zA-Z]:)?\\(?:[^\\/:*?"<>|\r\n]+\\)*[^\\/:*?"<>|\r\n]+\.(?:png|jpg|jpeg|gif)'
        image_content = re.findall(image_pattern, post)

        ## removing image path from text content
        for image_path in image_content:
            text_content = text_content.replace(image_path, "")

        ## extracting emojis
        emoji_set = demoji.findall(post)
        emojis = list(emoji_set.keys())

        ## extracting hashtags
        hashtag_pattern = r"\B#\w*[a-zA-Z]+\w*"
        hashtags = re.findall(hashtag_pattern, post)

    return text_content.strip(), links, image_content, hashtags, emojis
```

```python
df[['Text Content', 'Links', 'Image Content', 'Hashtags', 'Emojis']] = df['Post'].apply(extract_post_components).apply(pd.Series)
df
```

```python
df.to_csv('output.csv', index=False)
```
